feature3 = np.random.normal(0, 1, num_samples)

    'Feature 2': feature2,
    'Feature 3': feature3,

plt.title('Correlation Matrix Heatmap')
plt.xticks(rotation=45)
plt.yticks(rotation=45)

plt.show()

2. **Correlation Matrix Calculation**:

#### Ans.Here's a simplified example of a 3D surface plot for a basic function, such as 𝑧=𝑥2+𝑦2:

````python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Generate data
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
x, y = np.meshgrid(x, y)
z = x**2 + y**2  # Simple function

# Create a 3D surface plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
surface = ax.plot_surface(x, y, z, cmap='coolwarm')

# Add color bar, labels, and title
fig.colorbar(surface, ax=ax, shrink=0.5, aspect=5)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title(r'$z = x^2 + y^2$')

plt.show()
````

### 1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.

#### Ans.The Bokeh plot of a sine wave has been generated. It displays a sine curve with the x-values ranging from 0 to 10 and the y-values representing the sine of those x-values

````python
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import ColumnDataSource
import numpy as np

# Prepare the output for notebook
output_notebook()

# Step 1: Generate data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Step 2: Create a ColumnDataSource
source = ColumnDataSource(data=dict(x=x, y=y))

# Step 3: Create a figure
p = figure(title="Sine Wave", x_axis_label="X", y_axis_label="Sine(X)", width=800, height=400)

# Step 4: Add a line plot
p.line('x', 'y', source=source, line_width=2, color='blue')

# Step 5: Display the plot
show(p)
````

````python
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import ColumnDataSource

# Step 1: Prepare the output for notebook
output_notebook()

# Step 2: Define the dataset
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

# Step 3: Create a ColumnDataSource
source = ColumnDataSource(data=dict(fruits=fruits, counts=counts))

# Step 4: Create a figure for the bar chart
p = figure(x_range=fruits, title="Fruit Counts", x_axis_label="Fruits", y_axis_label="Count", 
           height=400, width=600, toolbar_location=None, tools="")

# Step 5: Add vertical bars
p.vbar(x='fruits', top='counts', width=0.9, source=source, color="skyblue")

# Step 6: Customize chart
p.xgrid.grid_line_color = None
p.y_range.start = 0

# Step 7: Display the bar chart
show(p)
````

````python
from bokeh.plotting import figure, show, output_notebook
import numpy as np

# Step 1: Prepare the output for notebook
output_notebook()

# Step 2: Generate random data
data_hist = np.random.randn(1000)

# Step 3: Create histogram data
hist, edges = np.histogram(data_hist, bins=30)

# Step 4: Create a figure for the histogram
p = figure(title="Histogram of Random Data", 
           x_axis_label='Value', 
           y_axis_label='Frequency', 
           height=400, 
           width=600)

# Step 5: Add bars to the histogram
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color='navy', line_color='white')

# Step 6: Display the histogram
show(p)
````

````python
from bokeh.plotting import figure, show, output_notebook
from bokeh.transform import linear_cmap
from bokeh.models import ColorBar, ColumnDataSource
from bokeh.io import show
from bokeh.palettes import Viridis256
import numpy as np

# Step 1: Prepare the output for notebook
output_notebook()

# Step 2: Create the dataset
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Step 3: Prepare the data for the heatmap
heatmap_data = data_heatmap.flatten()
x_coords = xx.flatten()
y_coords = yy.flatten()

# Step 4: Create a ColumnDataSource
source = ColumnDataSource(data=dict(value=heatmap_data, x=x_coords, y=y_coords))

# Step 5: Create a figure for the heatmap
p = figure(title="Heatmap", 
           x_axis_label='X', 
           y_axis_label='Y', 
           width=600, 
           height=600, 
           tools="")

# Step 6: Create a color mapper
mapper = linear_cmap(field_name='value', palette=Viridis256, low=min(heatmap_data), high=max(heatmap_data))

# Step 7: Add squares to the heatmap
p.rect(x='x', y='y', width=0.1, height=0.1, source=source,
        line_color=None, fill_color=mapper)

# Step 8: Add a color bar
color_bar = ColorBar(color_mapper=mapper['transform'], location=(0, 0))
p.add_layout(color_bar, 'right')

# Step 9: Display the heatmap
show(p)
````
